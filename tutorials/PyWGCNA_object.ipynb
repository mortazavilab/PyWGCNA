### Read PyWGCNA object

```python
import PyWGCNA
pyWGCNA_5xFAD = PyWGCNA.readWGCNA("5xFAD.p")
```

### Raw expression data

The raw expression data store in `geneExpr` in AnnData format which rows/obs are samples/sample information and cols/var are genes/gene information.

```python
print(f"Raw expresion data along with information:\n {pyWGCNA_5xFAD.geneExpr}")
```

```python
print(f"Raw sample inforamtions:")
pyWGCNA_5xFAD.geneExpr.obs.head(5)
```

```python
print(f"Raw gene inforamtions:")
pyWGCNA_5xFAD.geneExpr.var.head(5)
```

### Processed expresion matrix

After you ran PyWGCNA, you can use `datExpr` to look into results. The structure is similar to `geneExpr` which is AnnData format which rows/obs are samples/sample information and cols/var are genes/gene information. The main differences is now we have all the modules information.

```python
print(f"Processed expresion data along with information:\n {pyWGCNA_5xFAD.datExpr}")
```

```python
print(f"Raw sample inforamtions:")
pyWGCNA_5xFAD.datExpr.obs.head(5)
```

```python
print(f"Raw gene inforamtions:")
pyWGCNA_5xFAD.datExpr.var.head(5)
```

### Updating gene information

if you want to update gene inforamtion, you can use [`updateGeneInfo()` function](https://mortazavilab.github.io/PyWGCNA/html/api.html?highlight=updategeneinfo#PyWGCNA.wgcna.WGCNA.updateGeneInfo).

**note**: For doing functional enrichment analysis, you should have your gene name as a column (named should be gene_name) in your gene information dataframe.

[`getGeneList()` function](https://mortazavilab.github.io/PyWGCNA/html/api.html?highlight=getgenelist#PyWGCNA.utils.getGeneList) is helpping you to get gene name and all the other information you want to have like gene biotype using [biomart from ensemble](https://www.ensembl.org/biomart/martview/925c00ece2a04471e061de6e720c5cc7).

```python
geneList = PyWGCNA.getGeneList(dataset='mmusculus_gene_ensembl',
                               attributes=['ensembl_gene_id', 
                                           'external_gene_name', 
                                           'gene_biotype'],
                               maps=['gene_id', 'gene_name', 'gene_biotype'])


pyWGCNA_5xFAD.updateGeneInfo(geneList)
```

### Updating sample information

You can also update your sample information by using [`updateSampleInfo()` function](https://mortazavilab.github.io/PyWGCNA/html/api.html?highlight=updatesampleinfo#PyWGCNA.wgcna.WGCNA.updateSampleInfo).

```python
pyWGCNA_5xFAD.updateSampleInfo(path='5xFAD_paper/sampleInfo.csv', sep=',')
```
